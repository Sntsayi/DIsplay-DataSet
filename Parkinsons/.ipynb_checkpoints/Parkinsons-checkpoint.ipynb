```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df  = pd.read_csv("ParkinsonsData.csv")
df.head()
```

```python
plt.xlabel("NHR")
plt.ylabel("HNR")
plt.scatter(df.NHR , df.HNR)
plt.show()
```

```python
plt.xlabel("D2")
plt.ylabel("PPE")
plt.plot(df.D2 , df.PPE)
plt.show()
```
