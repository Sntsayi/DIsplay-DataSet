```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("carEvaluation.csv")
df.head()
```

```python
plt.xlabel("buyin")
plt.ylabel("doors")
plt.scatter(df.buying, df.doors)
plt.show()
```

```python
plt.xlabel("persons")
plt.ylabel("doors")
plt.plot(df.persons, df.doors ,linestyle = 'dotted')
plt.show()
```
