```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("adult.csv")
df.head()
```

```python
plt.xlabel("age")
plt.ylabel("workclass")
plt.scatter(df.age, df.workclass)
plt.show()
```

```python
plt.xlabel("education-num")
plt.ylabel("fnlwgt")
plt.scatter(df["education-num"], df.fnlwgt)
plt.show()
```

```python
plt.xlabel("education")
plt.ylabel("occupation")
plt.plot(df["education"], df.occupation)
plt.show()
```
