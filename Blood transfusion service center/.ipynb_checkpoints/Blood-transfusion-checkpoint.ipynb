```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("Blood-transfusion.csv")
df.head()
```

```python
plt.xlabel("Recency")
plt.ylabel("Frequency")
plt.plot(df.Recency, df.Frequency)
plt.show()
```

```python
plt.xlabel("Time")
plt.ylabel("Monetary")
plt.plot(df.Time, df.Monetary)
plt.show()
```
